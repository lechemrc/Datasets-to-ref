### Merging Dataframes

```
# Setting each dataset to have a tag in each column to indicate which dataset 
# they came from in the merged set 

emissions.columns = [str(col) + '_emissions' for col in emissions.columns]
emissions.columns.values[0] = 'country'

stringency.columns = [str(col) + '_stringency' for col in stringency.columns]
stringency.columns.values[0] = 'country'

num_threat['total'] = num_threat.sum(axis=1)
num_threat.columns = [str(col) + '_threatened' for col in num_threat.columns]
num_threat.columns.values[0] = 'country'

endangered['total'] = endangered.sum(axis=1)
endangered.columns = [str(col) + '_endangered' for col in endangered.columns]
endangered.columns.values[0] = 'country'

land_use['total'] = land_use.sum(axis=1)
land_use.columns = [str(col) + '_land_use' for col in land_use.columns]
land_use.columns.values[0] = 'country'

particulates.columns = [str(col) + '_particulates' for col in particulates.columns]
particulates.columns.values[0] = 'country'

lead.columns = [str(col) + '_lead' for col in lead.columns]
lead.columns.values[0] = 'country'

percent_species.columns = [str(col) + '_percent_species' for col in percent_species.columns]
percent_species.columns.values[0] = 'country'

population.columns = [str(col) + '_population' for col in population.columns]
population.columns.values[0] = 'country'

# Merging the dataframes
merged = emissions.merge(
    stringency, how='left').merge(
        num_threat, how='left').merge(
            endangered, how='left').merge(
                land_use, how='left').merge(
                    particulates, how='left').merge(
                        lead, how='left').merge(
                            percent_species, how='left').merge(
                                population, how='left')
```

### Regression

Since I will essentially be predicting the number of endangered species based on pollution, pollutants, and population, I will be running a regression to understand the correlations better. 

#### Train, Test, Val split


```
def train_validate_test_split(df, train_percent=.6, validate_percent=.2, seed=42):
    np.random.seed(42)
    perm = np.random.permutation(df.index)
    m = len(df.index)
    train_end = int(train_percent * m)
    validate_end = int(validate_percent * m) + train_end
    train = df.iloc[perm[:train_end]]
    validate = df.iloc[perm[train_end:validate_end]]
    test = df.iloc[perm[validate_end:]]
    return train, validate, test

train, val, test = train_validate_test_split(merged)

train.shape, val.shape, test.shape
```
